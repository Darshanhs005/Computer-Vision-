##Part 2 Feature based Classifiers

#a)
Generate a bag of features, containing
- Colour moment features including mean, standard deviation and skewness; [1.5 Marks]
- Grey-Level Co-occurrence Matrix (GLCM) based features of 'contrast', 'dissimilarity',
'homogeneity', 'ASM', 'energy', 'correlation'. [1.5 Marks]


```python
def normalise(feature):
    feature = np.array(feature)
    return (feature-feature.min())/(feature.max()-feature.min())
```

```python
def image_patches(img, p_size=(32, 32)):
    '''
    Splits img up into equal sized, non-overlapping patches.
    Ignores partial patches at the edge of the img.
    '''
    rows = img.shape[0]//p_size[0]
    cols = img.shape[1]//p_size[1]
    blocks = []
    for r in range(rows):
        for c in range(cols):
            y_min = r*p_size[0]
            y_max = y_min+p_size[0]
            x_min = c*p_size[1]
            x_max = x_min+p_size[1]
            b = img[y_min:y_max, x_min:x_max]
            blocks.append(b)
    return np.array(blocks).reshape((rows, cols, *p_size, -1))
```

```python

green_means = []
blue_means = []
red_means = []
green_stds = []
blue_stds = []
red_stds = []
dissimilarities = []
homogeneities = []
contrasts = []
ASMs = []
energys = [] 
correlations = [] 
skew_img_mean= []
skew_img_std= []
N_IMGS = len(dataset)
# Collect all features
for i in range(N_IMGS):
    img, lbl,_ = dataset[i]
    img = np.array(img)
    img_ori = img
    img = image_patches(img_ori)
    
    # Colour moment features
    r, g, b = img.mean(axis=(2,3))[0][0]
    # r, g, b = img.mean(axis=(0,1))
    red_means.append(r)
    green_means.append(g)
    blue_means.append(b)
    r, g, b = img.mean(axis=(2,3))[0][0]
    # r, g, b = img.mean(axis=(0,1))
    red_stds.append(r)
    green_stds.append(g)
    blue_stds.append(b)

    sk = skew(img_ori)
  
    skew_img_mean.append(sk.mean(axis=(0,1)))
    skew_img_std.append(sk.std(axis=(0,1)))
    # GLCM features
    img_gray = cv2.cvtColor(img_ori, cv2.COLOR_BGR2GRAY)
    
    cov = skimage.feature.greycomatrix(img_gray, distances=[1, 3, 5], angles=[0, math.pi/2, math.pi])
    
    contrasts.append(skimage.feature.greycoprops(cov, 'contrast'))
    dissimilarities.append(skimage.feature.greycoprops(cov, 'dissimilarity'))
    homogeneities.append(skimage.feature.greycoprops(cov, 'homogeneity'))
    ASMs.append(skimage.feature.greycoprops(cov, 'ASM'))
    energys.append(skimage.feature.greycoprops(cov, 'energy'))
    correlations.append(skimage.feature.greycoprops(cov, 'correlation'))
    # hog_f = skimage.feature.hog(img_ori)
    # hogs.append(hog_f)
    
    

green_means = normalise(green_means)
blue_means = normalise(blue_means)
red_means = normalise(red_means)
green_stds = normalise(green_stds)
blue_stds = normalise(blue_stds)
red_stds = normalise(red_stds)
skew_img_mean = normalise(skew_img_mean)
skew_img_std = normalise(skew_img_std)



contrasts = normalise(contrasts)
dissimilarities = normalise(dissimilarities)
homogeneities = normalise(homogeneities)
ASMs = normalise(ASMs)
energys = normalise(energys)
correlations = normalise(correlations)
# hogs = np.array(hogs)


# For the following, there are many features created per image. 
# Currently contrasts[i] is the vector of contrast features of image i.
# We want contrasts[i] to be the i-th contrast feature, over all images.
contrasts = contrasts.reshape([N_IMGS, -1]).transpose()
dissimilarities = dissimilarities.reshape([N_IMGS, -1]).transpose()
homogeneities = homogeneities.reshape([N_IMGS, -1]).transpose()
ASMs = ASMs.reshape([N_IMGS, -1]).transpose()
energys =energys.reshape([N_IMGS, -1]).transpose()
correlations =correlations.reshape([N_IMGS, -1]).transpose()
```

```python
def classification_accuracy(predictions, lbls):  
    return sum(1 for a, b in zip(predictions, lbls) if a == b)/len(predictions)

```

```python
features = np.stack([
    green_means,
    green_stds,
    blue_means,
    blue_stds,
    red_means,
    red_stds,
    skew_img_mean,
    skew_img_std,
    *contrasts,
    *dissimilarities,
    *homogeneities,
    *ASMs,
    *energys,
    *correlations,
  
], axis=-1)
print(features.shape)
```

```python
import torchvision.transforms.functional as tvtf
def preprocess_image(image):
    # Convert image to tensor
    image = tvtf.to_tensor(image)
    
    # Resize to 256px on shortest edge
    image = tvtf.resize(image, 256)
    
    # 224px centre crop
    image = tvtf.center_crop(image, 224)
    
    # Normalise the image
    image = tvtf.normalize(image, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    
    # Create a batch dimension
    image = image.unsqueeze(dim=0)
    
    return image
```

```python
def classify_image(model, image):
    # Set model to eval mode
    model.eval()
    
    # Preprocess the image
    image = preprocess_image(image)
        
    # Perform the forward pass
    with torch.no_grad():
        outputs = model(image)
    
    # Get the output vector, class index of highest score and class label
    output = outputs[0]                             # Extract the 1000D vector
    class_index = output.argmax(dim=0).item() 
       # Output is now a (1000,) dimensional vector, so we argmax over dim=0
    class_label = IMAGENET_CLASSES[class_index]
    
    # Convert the output vector to a numpy array
    output = output.detach().cpu().numpy()
    
    # Return the data
    return output, class_index, class_label
```
